```python
import pandas as pd 
import glob

from JSONreader import *
```

```python
path ='../webhose_data'

folders = [x[0] for x in os.walk(path)][1:]

def import_json_from_folder(folder):
    allFiles = glob.glob(folder + "/*.json")
    list_ = []
    for file_ in allFiles:
        df = json_to_dataframe(file_)
        list_.append(df)
    return pd.concat(list_).reset_index(drop=True)

def import_all_json_from_folders(folder_list):
    list_ = []
    for f in folder_list:
        df = import_json_from_folder(f)
        list_.append(df)
    return pd.concat(list_).reset_index(drop=True)
    
```

```python
frame = import_all_json_from_folders(folders)
# save the frame 
frame.to_csv('../Datasets/articles.csv', sep='\t', encoding='utf-8')
```
